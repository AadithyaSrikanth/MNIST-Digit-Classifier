```python
import numpy as np
from numpy.random import randint
```

```python
import tensorflow as tf
(train_X, train_y), (test_X, test_y) = tf.keras.datasets.mnist.load_data() 
```

```python
tf.version.VERSION
```

```python
train_X.shape
```

```python
train_X = train_X.reshape(train_X.shape[0],28,28,1).astype('float32')
test_X = test_X.reshape(test_X.shape[0],28,28,1).astype('float32')
input_shape = (28,28,1)

train_X = (train_X-127.5)/127.5
test_X  = (test_X-127.5)/127.5

print('train_X shape',train_X.shape)
print('Number of images in train_X',train_X.shape[0])
print('Number of images in test_X ',test_X.shape[0])
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D, Input
from tensorflow.keras.utils import plot_model

init = tf.random_normal_initializer(stddev=0.02)

input = Input(shape = input_shape)

x = (Conv2D(28, kernel_size=(3,3), input_shape = input_shape))(input)
x = (MaxPooling2D(pool_size = (2,2)))(x)
x = (Flatten())(x)
x = (Dense(128, activation = 'relu'))(x)
x = (Dropout(0.2))(x)
x = (Dense(10,activation= 'softmax'))(x)

model = Model(input,x)
```

```python
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
model.fit(x=train_X,y=train_y, epochs=20)
```

```python
model.evaluate(test_X, test_y)
```
